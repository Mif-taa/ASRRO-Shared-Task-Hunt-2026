# **Import Library**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Read the CSV File**

```python
df = pd.read_csv('/kaggle/input/customer-shopping-latest-trends-dataset/shopping_trends.csv')
```

# **Display the First Few Rows**

```python
df.head()
```

# **Basic overview**

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

# **Gender distribution**

```python
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.show()
```

# **Age distribution**

```python
plt.figure(figsize=(5,4))
sns.histplot(df['Age'], color='skyblue')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# **Average Purchase Amount by Gender**

```python
sns.barplot(data=df, x='Gender', y='Purchase Amount (USD)', estimator=np.mean)
plt.title('Average Purchase Amount by Gender')
plt.show()
```

# **Purchase Amount vs. Age**

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Age', y='Purchase Amount (USD)', hue='Gender')
plt.title('Purchase Amount vs Age')
plt.show()
```

# **Top 5 Popular Product Categories**

```python
top_categories = df['Category'].value_counts().nlargest(5)

plt.figure(figsize=(5,4))
sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")
plt.title('Top 5 Product Categories')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

# **Numpy Operations**

```python
amounts = df['Purchase Amount (USD)'].values

print("Max:", np.max(amounts))
print("Min:", np.min(amounts))
print("Mean:", np.mean(amounts))
print("Std Dev:", np.std(amounts))
```

# **Check for missing values**

```python
print(df.isnull().sum())
```

# **Encode categorical values**

```python
df['GenderEncoded'] = df['Gender'].map({'Female': 0, 'Male': 1})
```

```python
df.head()
```

# **Heatmap of correlations**

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# **Boxplot: Purchase by Category**

```python
plt.figure(figsize=(4,5))
sns.boxplot(data=df, x='Category', y='Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.title('Purchase Amount by Product Category')
plt.show()
```

# **Payment Method Distribution**

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Payment Method', order=df['Payment Method'].value_counts().index)
plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# **Purchases by Location**

```python
location_counts = df['Location'].value_counts().head(10)

plt.figure(figsize=(4,3))
sns.barplot(x=location_counts.index, y=location_counts.values, palette='coolwarm')
plt.title('Top 10 Locations by Purchase Count')
plt.xticks(rotation=45)
plt.ylabel('Number of Purchases')
plt.show()
```

# **Most Popular Product Colors**

```python
plt.figure(figsize=(4,4))
sns.countplot(data=df, y='Color', order=df['Color'].value_counts().head(10).index, palette='pastel')
plt.title('Top 10 Product Colors')
plt.xlabel('Count')
plt.ylabel('Color')
plt.show()
```

# **Seasonality in Shopping**

```python
plt.figure(figsize=(4,3))
sns.countplot(data=df, x='Season', order=['Spring', 'Summer', 'Autumn', 'Winter'])
plt.title('Purchases by Season')
plt.ylabel('Count')
plt.show()
```

# **Promo Code Usage by Season**

```python
plt.figure(figsize=(4,3))
sns.countplot(data=df, x='Season', hue='Promo Code Used')
plt.title('Promo Code Usage by Season')
plt.ylabel('Count')
plt.show()
```

# ***Cross-tab: Location vs Payment Method***

```python
location_payment = pd.crosstab(df['Location'], df['Payment Method'])

plt.figure(figsize=(12,6))
sns.heatmap(location_payment.head(10), cmap='YlGnBu', annot=True, fmt='d')
plt.title('Top 10 Locations vs Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Location')
plt.show()
```
